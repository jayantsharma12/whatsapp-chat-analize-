```python
chat_data_path = '/content/WhatsApp Chat with IBM ASSESSMENT .txt'

# Open and read the chat data file
with open(chat_data_path, 'r', encoding='utf-8') as file:
    chat_lines = file.readlines()

# data processing and sentiment analysis.
import spacy
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load spaCy for text processing
nlp = spacy.load("en_core_web_sm")

# Load the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Initialize lists to store message text and sentiment scores
messages = []
sentiment_scores = []

# Iterate through chat lines
for line in chat_lines:
    # Split line into sender and message text if ":" is present
    if ":" in line:
        sender, message = line.split(" ", 1)
        # Analyze sentiment using VADER
        sentiment = analyzer.polarity_scores(message)
        sentiment_score = sentiment['compound']
        messages.append(message.strip())
        sentiment_scores.append(sentiment_score)
    else:
        print(f"Ignoring line: {line}")

# Create a DataFrame to store message text and sentiment scores
data = {'Message': messages, 'Sentiment Score': sentiment_scores}
df = pd.DataFrame(data)
# sum=sum(sentiment_scores)
sum=0
for i in sentiment_scores:
  sum+=i
print(sum)
# print(sentiment_scores)
if sum>0:
  print("Positive")
elif sum<0:
  print("Negative")
else:
  print("Neutral")
df.head(46)

```
